```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException
import undetected_chromedriver as uc

# Set up Chrome options
options = webdriver.ChromeOptions()
options.headless = True
options.add_argument("start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

# Initialize the undetected Chrome driver
driver = uc.Chrome()

# Target website
website = 'https://www.thomasnet.com/suppliers/usa/food-trays-87561205?coverage_area=NA'
driver.get(website)

# Initialize lists to store scraped data
names = []
contact_no = []

def get_company_data():
    """Extracts company data (name and contact info) on the current page."""
    company_links = driver.find_elements(By.CSS_SELECTOR, "a[role='link']")
    for link in company_links:
        if link.is_displayed() and link.is_enabled():
            try:
                driver.execute_script("arguments[0].scrollIntoView(true);", link)
                time.sleep(1)
                link.click()

                # Extract company name
                try:
                    company_name = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.XPATH, '//*[@id="tnet_content"]/div[1]/div[2]/h1/a'))
                    ).text
                except TimeoutException:
                    print("Timeout while waiting for the company name.")
                    driver.back()
                    continue

                # Extract contact info
                try:
                    contact_info_element = driver.find_element(By.XPATH, '//*[@id="tnet_content"]/div[1]/div[2]/b/span')
                    contact_info = contact_info_element.text if contact_info_element else "N/A"
                except NoSuchElementException:
                    contact_info = "N/A"

                # Append data
                names.append(company_name)
                contact_no.append(contact_info)
            except (ElementClickInterceptedException, ElementNotInteractableException):
                print("Error clicking the link. Skipping...")
            finally:
                driver.back()
                time.sleep(2)

# Scrape multiple pages
for page_num in range(1, 12):  # Adjust as needed
    print(f"Scraping page {page_num}...")
    get_company_data()

    # Navigate to the next page
    try:
        next_button = driver.find_element(By.XPATH, '//a[contains(text(), "Next")]')
        driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
        next_button.click()
        time.sleep(3)
    except NoSuchElementException:
        print("No 'Next' button found. Ending pagination.")
        break

# Save the data
data = pd.DataFrame({'Company Name': names, 'Contact Info': contact_no})
data.to_csv('company_contact_trays.csv', index=False)

driver.quit()
print("Scraping completed successfully!")
```
