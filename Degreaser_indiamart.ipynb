```python
from requests_html import HTMLSession
import time
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
}

session = HTMLSession()
time.sleep(2)
r = session.get('https://dir.indiamart.com/impcat/water-based-degreaser.html', headers=headers)
print(r.status_code)
prod_name = []
company = []
product_cost = []
comp_address = []
Quantity = []

# Check if the initial request was successful
if r.status_code == 200:
    while True:
        r.html
        links = r.html.find('[href*=proddetail]')
        full_links = [item.attrs['href'] for item in links]
        print(full_links)

        for link in full_links:
            time.sleep(2)
            resp = session.get(link, headers=headers)

            # Check if the request was successful
            if resp.status_code == 200:
                resp.html
                try:
                    product_name = resp.html.find('h1.bo.center-heading')
                    for prod in product_name:
                        print(prod.text)
                        prod_name.append(prod.text)
                except:
                    prod_name.append("NA")

                try:
                    company_name = resp.html.find('a.color6.bo.atxu h2.fs15.bo')
                    for name in company_name:
                        print(name.text)
                        company.append(name.text)
                except:
                    company.append("NA")

                try:
                    prod_cost = resp.html.find('.prc-tip.bo.price-unit')  # Fix the CSS selector
                    for cost in prod_cost:
                        print(cost.text)
                        product_cost.append(cost.text)
                except:
                    product_cost.append("NA")

                try:
                    quantity = resp.html.find('span.fs15.pcl76', first=True)
                    if quantity is not None:
                        print(quantity.text)
                        Quantity.append(quantity.text)
                    else:
                        Quantity.append("NA")
                except Exception as e:
                    Quantity.append("NA")
                    print(e)

                try:
                    company_address = resp.html.find('div.fs12.color1.dsf.addrs.plhn')
                    for comp_add in company_address:
                        print(comp_add.text)
                        comp_address.append(comp_add.text)

                except:
                    comp_address.append("NA")

            else:
                print(f"Failed to fetch data from {link}. Status code: {resp.status_code}")

        # Find the "Show More Results" element
        show_more = r.html.find('span.fs15.br5.bg3.cp', first=True)

        if show_more is not None and "Show More Results" in show_more.text:
            # Click on "Show More Results" to load more pages
            show_more_link = show_more.find('a', first=True)
            if show_more_link is not None:
                next_page_url = show_more_link.attrs['href']
                r = session.get(next_page_url, headers=headers)
                print(next_page_url)
                continue
        else:
            break

    # Ensure all lists have the same length by filling missing data with "NA"
    max_length = max(len(prod_name), len(company), len(product_cost), len(comp_address), len(Quantity))
    prod_name += ["NA"] * (max_length - len(prod_name))
    company += ["NA"] * (max_length - len(company))
    product_cost += ["NA"] * (max_length - len(product_cost))
    comp_address += ["NA"] * (max_length - len(comp_address))
    Quantity += ["NA"] * (max_length - len(Quantity))

    # Create a DataFrame from the extracted data
    data = {
        'Product Name': prod_name,
        'Company Name': company,
        'Product Cost': product_cost,
        'Company Address': comp_address,
        'Quantity': Quantity
    }

    df = pd.DataFrame(data)
    print(df)

    # Export the DataFrame to an Excel file
#     df.to_excel('waterdegreaser_data_Indiamart_moreResults.xlsx', index=False)

else:
    print("Failed to fetch the initial page. Status code:", r.status_code)
```
